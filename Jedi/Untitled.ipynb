### Tensors

Definition from Wikipedia (pizdet):

<i>În matematică, un tensor este un obiect geometric care asociază într-o manieră multi-liniară vectori geometrici, scalari și alți tensori cu un tensor rezultat.</i>

In viziunea mea: <br>
Un tensor este un obiect geometric, care poate fi inteles ca o structura de date, si se ocupa cu stocarea datelor. Un tensor poate fi descris prin cateva caracteristici:

1. Numarul de elemente (size):<br>
Reprezinta numarul de elemente totale. De exemplu 60 de elemente
2. Numarul de dimensiuni (ndim):<br>
Reprezinta gradul tensorului. Un tensor de 0 dimensiuni este numit scalar (de exemplu 7). un tensor de 1 dimensiuni este numit vector (de exemplu \[0, 2\]). Un tensor de 2 dimensiuni este numit matrice (de exemplu \[\[0,1\]\[1,0\]\]), etc...
3. Shape (Forma):<br>
Reprezinta numarul de elemente care contine fiecare dimensiune.

<i>Exemple</i>

```python
scalar = np.array(1)
print(f'Tensor - {scalar}.')
print(f'Size - {scalar.size}, Ndim - {scalar.ndim}, Shape - {scalar.shape}')
```

```python
vector = np.array([1,2])
print(f'Tensor - {vector}.')
print(f'Size - {vector.size}, Ndim - {vector.ndim}, Shape - {vector.shape}')
```

```python
matrix = np.array([[0,1],[1,0]])
print(f'Tensor - {matrix}.')
print(f'Size - {matrix.size}, Ndim - {matrix.ndim}, Shape - {matrix.shape}')
```

### Derivatives

From wikipedia:<br>
<i>Derivata unei funcții într-un punct semnifică rata cu care se modifică valoarea funcției atunci când se modifică argumentul. Cu alte cuvinte, derivata este o formulare matematică a noțiunii de rată de variație.</i>

Derivata este o functie, obtinuta prin efectuarea unor operatii asupra altei functii, care ne indica viteza de crestere/descrestere intr-un anumit punct al functiei.

De exemplu, in imaginea de mai jos avem functia f(x)= x^3 si functia ei derivata f'(x) = 3x^2


Sa analizam punctul x0 = -1. <br>
Inlocuind f'(x0) vom primi f'(-1) = 3*(-1)^2 = 3. <br>
3 este un numar pozitiv. Acest fapt presupune ca in punctul x = -1, functia f(x)=x^3 va fi crescatoare.

Derivata, pe langa asta ne permite sa calculam punctele de minim si maxim (globale si locale). <br>
Insa acest lucru devine destul de greu pentru functiile de mai multe variabile, de asta el este aplicat in mare parte doar pe functiile de o variabila, ca cea indicata in exemplul de mai sus.

Derivarea functiilor de mai multe variabile are loc in mare parte la fel. Insa o trasatura deosebita este ca in urma derivarii nu primim o singura functie derivata, ci mai multe, numarul acestora fiind egala cu numarul de variabile.<br>

De exemplu:<br>
f(x,y) = x^2 + y^2<br><br>

derivata dupa x:<br>
d/dx = 2x + 0<br><br>

derivata dupa y:<br>
d/dy = 0 + 2y

Gradientul este vectorul al carui elemente sunt derivatele functiei.<br>

De exemplu pentru f(x,y) gradientul va fi (d/dx, d/dy)
