```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/Churn.csv')
df
```

```python
df.info()
```

```python
# Stergem coloana id, care nu prezinta o informație relevantă
df.drop(columns=['customerID'], inplace = True)
df
```

```python
# Transformam coloana TotalCharges din object(string) in float
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.info()
```

```python
# Controlam Randurile cu Valorile NaN
df[df['TotalCharges'].isnull()]
```

```python
# Inlocuim Valorile NaN cu valoarea medie
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
df
```

```python
# Transformam valorile string in categorii, iar apoi in valori int
# Alcatuim un dictioniar instructiune

settings = dict()
for column in df.columns:
    count = 0
    if df[column].dtype in ['object', 'category']:
        if df[column].dtype == 'object':
            df[column] = pd.Categorical(df[column])
        for key in df[column].value_counts().keys():
            settings[column + '_' + key] = count
            count += 1
        df[column] = df[column].cat.codes
```

```python
# Controlam corelatia dintre coloane mai mare ca 0.80
data = df.corr()
data = data[abs(data) > 0.80].stack().reset_index()
data[data['level_0'] != data['level_1']]
```

```python
# Controlam corelatia dintre features cu target mai mica ca 0.01
data = df.corr()
data = data[abs(data) < 0.01].stack().reset_index()
data[data['level_0'] == 'Churn']
```

```python
df = df.drop(columns=['gender', 'TotalCharges'])
df
```

```python
plt.figure(figsize = (25, 25))
count = 1

for column in df.columns:
    plt.subplot(5, 4, count)
    sns.distplot(df[column])
    plt.xlabel(column)
    count += 1
    if count == 18: break

plt.show()
```

```python
from sklearn.model_selection import train_test_split

x = df[df.columns[:-1]].values
y = df[df.columns[-1]].values
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 16)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression(C=8, max_iter = 1000)
model.fit(X_train, y_train)
print(accuracy_score(model.predict(X_test), y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 20, p=3)
model.fit(X_train, y_train)
print(accuracy_score(model.predict(X_test), y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(min_samples_leaf=40)
model.fit(X_train, y_train)
print(accuracy_score(model.predict(X_test), y_test))
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(C = 20)
model.fit(X_train, y_train)
print(accuracy_score(model.predict(X_test), y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
print(accuracy_score(model.predict(X_test), y_test))
```

```python
model = CategoricalNB()
model.fit(X_train, y_train)
print(accuracy_score(model.predict(X_test), y_test))
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
print(accuracy_score(model.predict(X_test), y_test))
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_leaf=5, max_depth=25)
model.fit(X_train, y_train)
print(accuracy_score(model.predict(X_test), y_test))
```
