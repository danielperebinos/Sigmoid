# Clustering

### Reading Data

```python
df = pd.read_csv('Wine.csv')
df
```

```python
X = df[df.columns[:-1]]
y = df[df.columns[-1]]
```

<br><br>

### KMeans Algorithm

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
```

```python
range_n_clusters = range(2, 10)

for n_clusters in range_n_clusters:

    fig, ax = plt.subplots()
    fig.set_size_inches(10, 5)

    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    silhouette_avg = silhouette_score(X, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )

    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax.scatter(
        X[df.columns[1]], X[df.columns[12]], marker=".", s=200, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )


    centers = clusterer.cluster_centers_
    ax.scatter( centers[:, 1], centers[:, 12],
        c='red' , s=200
    )

    for i, c in enumerate(centers):
        ax.scatter(c[0], c[1], marker=f"${i}$", s=50, edgecolor="k")

    ax.set_title("The visualization of the clustered data.")
    ax.set_xlabel("Feature space for the 1st feature")
    ax.set_ylabel("Feature space for the 12nd feature")

    plt.suptitle(
        f"Silhouette analysis for KMeans clustering on sample data with n_clusters = {n_clusters}",
        fontsize=14,
        fontweight="bold",
    )

plt.show()
```

```python
clusterer = KMeans(n_clusters=2, random_state=10)
cluster_labels = clusterer.fit_predict(X)
```

### Gaussian Mixture (GMM) Algorithm

```python
from sklearn.mixture import GaussianMixture
from matplotlib.patches import Ellipse
```

```python
range_n_clusters = range(2, 10)

for n_clusters in range_n_clusters:

    fig, ax = plt.subplots()
    fig.set_size_inches(10, 5)
    
    gmm = GaussianMixture(n_components=n_clusters).fit(X)
    labels = gmm.predict(X)
    
    silhouette_avg = silhouette_score(X, labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )
    
    ax.scatter(X[df.columns[0]], X[df.columns[12]], c = labels, s = 200, cmap='viridis')
    ax.set_title("The visualization of the clustered data.")
    plt.suptitle(
        f"Silhouette analysis for GMM clustering on sample data with n_clusters = {n_clusters}",
        fontsize=14,
        fontweight="bold",
    )
```
