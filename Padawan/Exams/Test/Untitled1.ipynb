```python
import requests
import bs4
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_car_brands'

req = requests.get(url)
req
```

```python
soup = bs4.BeautifulSoup(req.text, 'html.parser')
```

```python
soup.prettify().find('Mitsubishi')
```
