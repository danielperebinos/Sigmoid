```python
import numpy as np
import pandas as pd
from matplotlib import patches, pyplot as plt
import seaborn as sns
```

<br><br>

### Normal Distribution

```python
normal = np.random.normal(0, 1, (10000, 2))
normal.shape
```

```python
plt.figure(figsize=(20,10))
sns.set_style('dark')
sns.histplot(normal, bins=50, kde=True)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.hist2d(normal[:, 0], normal[:, 1])
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(normal[:, 0], normal[:, 1], c='purple')
plt.show()
```

### Uniform Distribution

### Poisson Distribution

```python
poisson = np.random.poisson(lam=10, size=(1000,2))
poisson.shape
```

```python
plt.figure(figsize=(12,10))
plt.hist2d(poisson[:, 0], poisson[:, 1])
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(poisson[:, 0], poisson[:, 1], c='green')
plt.show()
```
